Feature engineering is the process of using domainknowledge to extract features from raw data . These features can be used to improve the perfromance of ML algos. 
good data >>>>> good algo.

Feature engineeering is of 4 kinds
A. Feature transformation (feature = column)
   - missing value imputation
   - handling categorical features
   - outlier detection
   - feature scaling.

B. feature construction 
C. feature selection 
D. Feature extraction
