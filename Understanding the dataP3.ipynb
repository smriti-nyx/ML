1. Scatterplot(numerical - numerical)


```python
sns.scatterplot(y= tips['tip'] , x= tips['total_bill'])
```

```python
#multivariate analysis 
```

```python
sns.scatterplot(y= tips['tip'] , x= tips['total_bill'], hue = tips['sex']#)
```

```python
#adding one more parameter 
sns.scatterplot(y= tips['tip'] , x= tips['total_bill'] , hue = tips['sex'], style = tips['smoker'])
```

```python
#adding one more parameter
sns.scatterplot(y= tips['tip'] , x= tips['total_bill'] , hue = tips['sex'] , style = tips['smoker'] , size = tips['size'])
#how many people were brought in each group is represented by size . 
```

4. Distplot (num/cat)

```python
sns.distplot(titanic[titanic['Survived'] == 0 ]['Age'], hist = False) #dead, blue
sns.distplot(titanic[titanic['Survived'] == 1 ]['Age'], hist = False) #survived , orange
 
```

5. Heatmap(cat/cat)

```python
titanic.head()
```

```python
#printing out a crosstab
sns.heatmap(pd.crosstab(titanic['Pclass'] , titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean(numeric_only = True )['Survived']*100).plot(kind = 'bar')
```
