EDA (Univariate)

analysis of single variable (column) is called univariate analysis .
for 2 col - bivariate
multiple cols - multivariate analysis .


there are 2 kinds of data - numerical or categorical.

```python
import pandas as pd
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

1.Categorical data


a.countplot

```python
import seaborn as sns
```

```python
#sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
df['Pclass'].value_counts().plot(kind = 'bar')
```

```python
df['Sex'].value_counts().plot(kind = 'bar')
```

b. pie chart

```python
df['Survived'].value_counts().plot(kind = 'pie')
```

```python
#to show percentage
df['Pclass'].value_counts().plot(kind = 'pie' , autopct = '%.2f' )
```

2. Numerical data

a. Histogram

```python
import matplotlib.pyplot  as plt
plt.hist(df['Age'] , bins = 40)
```

b. Displot

```python
sns.displot(df['Age'])
```

c. Boxplot

Helps us to identify the outliers . for instance the boxplot below has a passenger who paid 500 dollars as fare , so we cant rely on such kind of data.

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```

```python
df['Age'].max()
```

```python
df['Age'].min()
```

```python
#to find if the data is skewed or not
df['Age'].skew()
#if it is zero , means the data is symmetrical /skewed . The data curve peaks at x = 0 .
```
