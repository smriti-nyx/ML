1. CSV

```python
import pandas as pd
```

opening a csv file froom an url 

```python
import requests
from io import StringIO

url = " "
headers = { "User-Agent" : "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14 ; rv: 66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url , headers = headers)
data = StringIO(req.text)

pd.read_csv(data)
```

SEP parameter

```python
#separator of csv files . we can overwrite the separators
pd.read_csv(' ' , sep = '\t') #tab
```

```python
#if no column names have been given already , the first row elemnts take up the names of the column. To add column name :
pd.read_csv(' ' , sep = '\t' , names = 'X', 'Y', 'Z', 'W')
```

index_col parameter

```python
#to convert one of the given columns into the index column :
pd.read_csv(' name.csv' , index_col = 'name of that col') #tab
```

Header parameter 

```python
#the first row contains the col names and marked as row 0 and the first row is marked 1. To make those names the col  names :
pd.read_csv(' ' , header = 1) #tab

```

use_cols parameter

```python
#to see selective columns
pd.read_csv(' ' , usecols = ['col1', 'col2']) #tab
```

squeeze parameters 

```python
#to view only one column.
pd.read_csv(' ' , usecols = [ ''] , squeeze = True ) #tab
```

skip rows / nrows columns

```python
pd.read_csv(' name.csv' , skiprows = [ 0,1] )
            #skips the first and second row
```

encoding parameter

```python
#diff datasets are encoded differently , to fix xit
#1. change encoding  by going to the subline
#2. check the encoding and put it , so that the datset loads
pd.read_csv(' ' , encoding = ' ') #tab
```

skip bad lines

```python
#sometimes rows have more col values than rest of the rows , #parser error
#to fix it
pd.read_csv(' ' ,error_bad_lines = False ) #tab
```

dtypes parameter

```python
#to change the data type of a column
pd.read_csv(' ' ,dtype = { 'target' :int} ) #tab
```

handling dates

```python
#in pandas , string = object , if date is given in object format , fix it
pd.read_csv(' ' , parse_dates = [ 'col name containing date'])
#can also merge the date , month , year col althogether into one
```

convertors

```python
#to convert full names into short forms by creating a fnc
def rename(name):
    if name == 'royal challenger':
        return "RCB"
    else:
        return " "
```

```python
rename("royal challen....")
#it will be done
pd.read_csv(' ' , converters = { 'col_name : rename '}) #tab
```

na_values parameters

```python
#to identify nan/ hypehn(-) values , or more data which are to considered  as null values
pd.read_csv(' ' , na_values = ['the_sign'] )
```
