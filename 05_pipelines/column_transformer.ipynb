Column Transformer

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('usedcar.csv')
```

```python
df.drop(['kmDriven', 'PostedDate', 'AdditionInfo'], axis = 1)
```

```python
#with column transformer, we can perform all the diff encodings to the diff columns acc to their requirement in just one line.

#transmission & fuel type - nominal ohe
#owner = ordinal -ordinal encoding
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.drop('Year', axis=1),
    df['Year'],
    test_size=0.2,
    random_state=0
)
```

```python
transformer = ColumnTransformer(transformers =[
    ('tnf1' , SimpleImputer() , ['Age']) ,
    ('tnf2' , OrdinalEncoder(categories = [['first','second']]) , ['Owner']) ,
    ('tnf3' , OneHotEncoder(sparse_output = False , drop = 'first'), ['Transmission' , 'FuelType'])
] , remainder = 'passthrough')
```

```python
transformer.fit_transform(x_train).shape
```
