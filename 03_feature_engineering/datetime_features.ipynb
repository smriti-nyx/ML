**WORKING WITH DATES**

```python
#converting to date time data type
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract Year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract month

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_month'] = date['date'].dt.month_name()
date.head()
```

3. Extract days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
#day of the week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
#day of week - name
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns = ['product_id','city_id', 'orders']).head()
```

```python
#is weekend ?
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
date.drop(columns = ['product_id', 'city_id','orders']).head()
```

```python
#extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns = ['product_id', 'city_id','orders']).head()
```

```python
#extratc semester
date['quarter'] = date['date'].dt.quarter
date['semester']= np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns = ['product_id', 'city_id','orders']).head()
```

time elapsed bw two dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

months passed

```python
np.round((today-date['date'])/np.timedelta64(1, 'm'), 0)
```
